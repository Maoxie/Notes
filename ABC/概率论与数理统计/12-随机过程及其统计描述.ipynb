## 1. 随机过程的概念

**随机**：研究对象是随机的现象

**过程**：随时间演变

用一族（无限多个）随机变量来描述研究的对象。

### 定义：

设 $T$ 是一无限实数集。把依赖于参数 $t \in T$ 的一族（无限多个）随机变量称为**随机过程**，记为 $\{X(t),t\in T\}$，这里对每一个 $t \in T$，$X(t)$ 是一随机变量。 $T$叫做**参数集**。

$t$取离散整数值时，称为该随机过程为离散参数随机过程，记作 $\{X_n,n\in \mathbb{N}\}$；
$t$取连续值时，称为该随机过程为连续参数随机过程，记作 $\{X(t),t\in \mathbb{R}\}$。

通常把$t$看作时间，称$X(t)$为时刻$t$时过程的状态，而 $ X(t_1)=x $ （实数）说成是 $t=t_1$ 时过程处于状态 $x$。对于一切$t \in T$，$X(t)$所有可能取的一切值的全体称为随机过程的**状态空间**。

对随机过程进行一次试验，其结果 $x(t), t\in T$ 称为随机过程的一个**样本函数**或**样本曲线**。

#### 简单理解：

依赖于参数t的一组随机变量。

通常在不同时刻 $t_1, t_2, t_3, ...$ 时的随机样本 $x(t_1), x(t_2), x(t_3), ...$ 之间并不独立，因此不能简单当做多维随机变量来处理。

## 2. 随机过程的统计描述

### （一）随机过程的分布函数族

随机过程 $\{X(t),t\in T\}$，对于每一个固定的 $t\in T$，随机变量 $X(t)$ 的分布函数一般与 $t$ 有关，记为
$$
F_X(x,t)=P\{X(t)\le x\},x\in R
$$
称它为随机过程 $\{X(t),t\in T\}$ 的**一维分布函数**，而 $\{F_X(x,t), t \in T\}$称为**一维分布函数族**。

对于$n$个不同时刻，
$$
F_x(x_1,x_2,...,x_n;t_1,t_2,...,t_n)=P\{X(t_1\le x_1,X(t_2)\le x_2,...,X(t_n)\le x_n)\}, \\
x_i \in R, i=1,2,...,n
$$
$n$是常数，此时称$\{F_x(x_1,x_2,...,x_n;t_1,t_2,...,t_n),t_i \in T\}$ 为随机过程$\{X(t),t\in T\}$的**n维分布函数族**

### （二）随机过程的数字特征

#### 均值函数

$t$时刻$X(t)$是一随机变量，它的均值记为
$$
\mu_X(t)=E[X(t)]
$$
称$\mu_X(t)$为随机过程$\{X(t),t \in T\}$的**均值函数**

这是随机过程所有样本函数的$t$时刻的函数值的平均值，通常称为**集平均**或**统计平均**

均值函数表示了随机过程在各个时刻的摆动中心。

#### 均方值函数、方差函数、标准差函数

随机变量$X(t)$的二阶原点矩和二阶中心矩分布记为
$$
\Psi_X^2(t)=E[X^2(t)] \\
\sigma_X^2(t)=D_X(t)=Var[X(t)]=E\{[X(t)-\mu_X(t)]^2\}
$$
称为随机过程$\{X(t),t \in T\}$的**均方值函数**、**方差函数**。
方差函数的算数平方根$\sigma_X(t)$称为随机过程的**标准差函数**，它表示随机过程对于均值的平均偏离程度。



#### 自相关函数、自协方差函数

$X(t_1)$和$X(t_2)$的二阶原点混合矩记为
$$
R_{XX}(t_1,t_2)=E[X(t_1)X(t_2)]
$$
称为随机过程$\{X(t),t \in T\}$的**自相关函数**，简称**相关函数**。常简记为$R_X(t_1,t_2)$。

$X(t_1)$和$X(t_2)$的二阶混合中心矩记为
$$
\begin{aligned}
C_{XX}(t_1,t_2) &= Cov[X(t_1)X(t_2)] \\
&=E\{[X(t_1)-\mu_X(t_1)][X(t_2)-\mu_X(t_2)]\}
\end{aligned}
$$
称为随机过程$\{X(t),t \in T\}$的**自协方差函数**，简称**协方差函数**。常简记为$C_X(t_1,t_2)$。

自相关函数和自协方差函数是刻画随机过程自身在两个不同时刻的状态之间统计依赖关系的数字特征。

#### 数字特征之间的关系

$$
\Psi_X^2(t)=R_X(t,t) \\
C_X(t_1,t_2)=R_X(t_1,t_2)-\mu_X(t_1)\mu(t_2) \\
\sigma_X^2(t)=C_X(t,t)=R_X(t,t)-\mu_X^2(t)
$$
随机过程的数字特征中最主要的是均值函数和自相关函数

#### 正态过程

如果对每个$t \in T$，随机过程$\{X(t),t \in T\}$的二阶矩$E[X^2(t)]$都存在，则称它为**二阶矩过程**。

如果随机过程$\{X(t),t \in T\}$的每一个有限维分布都是正态分布，则称它为**正态过程**。即对任意正整数$n \ge 1$及任意$t_1,t_2,...,t_n \in T,(X(t_1\le x_1,X(t_2)\le x_2,...,X(t_n))$服从n维正态分布。

### （三）二维随机过程的分布函数和数字特征

设$X(t),Y(t)$是依赖于同一参数$t \in T$的随机过程，称 $\{(X(t),Y(t)),t \in T\}$为**二维随机过程**。

## 3 泊松过程及维纳过程

给定二阶矩过程$\{X(t), t \ge 0\}$，称随机变量$X(t)-X(s), 0\le s<t$为随机过程在区间$(s,t]$上的增量。

如果对任意正整数n和$0\le t_0<t_1<t_2<...<t_n$，n个增量
$$
X(t_1)-X(t_0),X(t_2)-X(t_1),...,X(t_n)-X(t_{n-1})
$$
相互独立，则称$\{X(t), t \ge 0\}$为**独立增量过程**。

“在互不重叠的区间上，状态的增量是相互独立的”

如果对于人实数$h$和$0\le s+h<t+h$，$X(t+h)-X(s+h)$与$X(t)-X(s)$具有相同的分布，则称**增量具有平稳性**。当增量具有平稳性时，称相应的独立增量过程是**齐次的**或**时齐的**

对于独立增量过程，记$Y(t)=X(t)-\mu_X(t)$，当$0\le s<t$时，协方差函数
$$
\begin{aligned}
C_X(s,t) &=E[Y(s)Y(t)] \\
&=E\{[Y(s)-Y(0)][(Y(t)-Y(s))+Y(s)]\} \\
&=E[Y(s)-Y(0)]E[Y(t)-Y(s)]+E[Y^2(s)] \\
&=D_X(s)
\end{aligned}
$$
于是，协方差函数可以用方差函数表示为
$$
C_X(s,t)=D_X(min\{s,t\})
$$

### （一）泊松过程

如果随机过程$N(t), t \ge 0$表示时间段$(0,t]$内发生某种事件的总数，则称随机过程$N(t)$为**计数过程**。

例如：在$[0,t]$之内到达某商店的顾客人数。

$N(t)-N(t_0)$表示从时刻$t_0$到时刻t之间发生的事件次数，记为$N(t_0,t)$，一般情况下$N(t_0,t)$为一随机变量。

如果计数过程$N(t)$满足下述条件：

1. $N(0)=0$；
2. $N(t)$是独立增量过程（即，在不重叠的区间上增量具有独立性）；
3. $N(t)$是平稳增量过程（即，$N(t,t+\Delta t)=N(0,\Delta t)=N(\Delta t)$）；
4. $$\frac{P(N(t+\Delta t)-N(t))\ge 2}{P(N(t+\Delta t)-N(t)) = 1} \to 0, \Delta t \to 0$$即，对于充分小的$\Delta t$，在$(t,t+\Delta t]$内出现2次或2次以上事件的概率与出现1次的概率相比可以忽略不计。

则称其为**泊松过程**。

记$P_0(s)=P(N(s)=0)$，基于上述性质，有$\forall t,s \ge 0$
$$
\begin{aligned}
P_0(t+s) &= P(N(t+s)=0)=P(N(s)=0,N(t+s)-N(s)=0) & \\
    &= P(N(s)=0)P(N(t+s)-N(s)=0) & （独立增量过程） \\
    &= P(N(s)=0)P(N(t)=0) & （平稳增量过程） \\
    &= P_0(s)P_0(t)
\end{aligned}
$$

考虑到$P_0(0)=P(N(0)=0)=1$，因此上述方程的解必定为以下形式
$$
P_0(t)=exp({-\lambda t})
$$
常数$\lambda \ge 0$称为泊松过程的**强度**。

#### 1. 泊松过程的分布律

(推导过程略）

$$
P(N(t)=k)=\frac{(\lambda t)^k}{k!}exp(-\lambda t)
$$

即，泊松过程的一维分布是泊松分布。

换个角度看：把$(0,t]$区间均匀地划分为n个相等的部分，设n非常大，则每个区间的长度都非常小，以至于每个小区间内发生2次及以上事件的概率都可以忽略（性质4），那么N(t)就是发生1次事件的区间的个数，它服从二项分布。根据泊松定理，$n \to \infty$时，该二项分布可以用泊松分布表示。

由分布律可知因此泊松过程的均值函数和方差函数分别为：

$$E[N(t)]=\lambda t$$
$$D_N(t)=\lambda t$$

#### 2. 等待时间 

设事件依次重复出现的时刻 **$t_1,t_2,...,t_n,...$** 是一强度为$\lambda$的泊松流，$N(t)$为相应的泊松过程。记
$$W_0=0,W_n=t_n,n=1,2,...$$
$W_n$表示第n次随机事件出现的**等待时间**。

$W_n$的概率分布函数为：
$$
\begin{aligned}
F_{W_n}(t) &= \sum_{k=n}^{+\infty}exp(-\lambda t)\frac{(\lambda t)^k}{k!}, t\ge 0 \\
F_{W_n}(t) &= 0, t<0
\end{aligned}
$$

$W_n$概率密度函数为：
$$
f_{W_n}(t)=\frac{dF_{w_n}(t)}{dt}=\left\{ \begin{aligned} &\frac{\lambda(\lambda t)^{n-1}}{(n-1)!}exp(-\lambda t), &t>0 \\
    & 0, &t \le 0 \\
\end{aligned}
\right.
$$
即，**泊松过程的等待时间服从$\Gamma$分布**。

特别的，令n=1，**泊松过程事件的首次出现的等待时间$W_1$服从指数分布**
$$
f_{W_n}(t)=\left\{ \begin{aligned} &\lambda e^{-\lambda t}, &t>0 \\
    & 0, &t \le 0 \\
\end{aligned}
\right.
$$

#### 3. 点间距离

记$$T_i=W_i-W_{i-1},i=1,2,...$$
称为第i-1个事件与第i个事件的**点间距离**

1. 强度为$\lambda$的泊松过程的点间距离是相互独立的随机变量，且服从同一个指数分布$$
f_{T_i}(t)=\left\{ \begin{aligned} &\lambda e^{-\lambda t}, &t>0 \\
    & 0, &t \le 0 \\
\end{aligned}
\right. i=2,3,...
$$

2. 如果任意相继出现的两个事件的点间距离是相互独立，且服从同一个指数分布，则它们构成了强度为$\lambda$的泊松过程。

### （二）维纳过程

维纳过程是布朗运动的数学模型。

对于二阶矩过程$\{W(t),t\ge 0\}$，如果它满足
1. 具有独立增量；
2. 对任意的$t>s\ge 0$，增量$$W(t)-W(s)\sim N(0,\delta^2(t-s)),\delta>0$$
3. $W(0)=0$
则称此过程为**维纳过程**。

由2可知，维纳过程的增量的分布只与时间差有关，所以它是齐次的独立增量过程，也是正态过程。

维纳过程的分布完全由它的均值函数和自协方差函数（或自相关函数）所确定。

$$W(t) \sim N(0,\delta^2t)$$

均值函数和方差函数分别为
$$
E[W(t)]=0, D_W(t)=\sigma^2t
$$

其中$\sigma^2$称为维纳过程的参数，可通过实验观察值加以估计。
